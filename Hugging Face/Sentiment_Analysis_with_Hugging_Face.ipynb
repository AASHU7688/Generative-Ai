**Import libraries**

```python
from transformers import pipeline
from datasets import load_dataset
```

 **Load Dataset**

```python
dataset = load_dataset("imdb")
```

```python
print(dataset)
```

```python
dataset["train"][0]
```

**Create Sentiment Analysis Model**

```python
sentiment_ai = pipeline("sentiment-analysis")
```

**Test with our own text**

```python
sentiment_ai("I really loved this movie")
```

**Run model on dataset samples**

```python
for i in range(5):
    text = dataset["test"][i]["text"]
    result = sentiment_ai(text)
    print(result)
```
